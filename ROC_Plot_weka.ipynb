```python
import pandas as pd
data1 = pd.read_csv('ROC_RF_final.csv')  
data2 = pd.read_csv('rdkit_kstar_roc.csv')  

```

```python
fpr1 = data1.iloc[:,1:2]
fpr2 = data2.iloc[:,1:2]
fpr1

```

```python
tpr1 = data1.iloc[:,0:1]
tpr2 = data2.iloc[:,0:1]
tpr2
```

```python
import matplotlib.pyplot as plt
plt. plot(fpr1, tpr1, color = 'green', label='Random forest_rdkit')
plt. plot(fpr2, tpr2, color = 'blue', label='kstar_rdkit')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```

```python
from sklearn.metrics import auc
roc_auc = auc(fpr1, tpr1)
print('ROC AUC:', roc_auc)
```

```python
roc_auc = auc(fpr2, tpr2)
print('ROC AUC:', roc_auc)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import auc

plt.plot(fpr1, tpr1, label='Randomforest_rdkit', alpha=0.9)
plt.plot(fpr2, tpr2, label='Kstar_rdkit')
plt.plot([0, 1], [0, 1], linestyle='--', color='black')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
auc1 = auc(fpr1, tpr1)
auc2 = auc(fpr2, tpr2)
plt.legend(['Randomforest_rdkit (AUC = %0.3f)' % auc1, 'Kstar_rdkit (AUC = %0.3f)' % auc2])
plt.show
plt.savefig("Features.jpg", bbox_inches = 'tight', dpi=1200)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

data3 = pd.read_csv('rf_PRC.csv')
precision1 = data3['Precision']
recall1 = data3['Recall']
thresholds1 = data3['Threshold']
label1 = 'Classifier 1'

data4 = pd.read_csv('Kstar_PRC.csv')
precision2 = data4['Precision']
recall2 = data4['Recall']
thresholds2 = data4['Threshold']
label2 = 'Classifier 1'


plt.plot(recall1, precision1, label=label1)
plt.plot(recall2, precision2, label=label2)

plt.plot(thresholds1, precision1, 'b--', label='Threshold 1')
plt.plot(thresholds1, recall1, 'g--', label='Threshold 1')
plt.plot(thresholds2, precision2, 'r--', label='Threshold 2')
plt.plot(thresholds2, recall2, 'c--', label='Threshold 2')

plt.xlabel('Threshold')
plt.ylabel('Precision/Recall')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```
